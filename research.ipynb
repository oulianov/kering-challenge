Goal : create a comprehensive sustainability index to guide consumers towards more sustainable products. 

```python
import pandas as pd

#df = pd.read_csv("data/epl-valued-results-2020.csv", sep=";")
df_country_ekpi = pd.read_csv("data/raw-material-intensities-2020.csv", sep=";")
df_country_ekpi.where(df_country_ekpi["Indicator unit"] == "kg")
```

```python
df_country_ekpi_agg = df_country_ekpi.groupby(["Material Group", "Impactcountry"]).sum().reset_index()
df_country_ekpi_agg = df_country_ekpi_agg.merge(
    df_country_ekpi.groupby("Material Group").sum()\
        .reset_index()[["Material Group", "Valued_Ekpi_intensity"]]\
            .rename({"Valued_Ekpi_intensity": "Material Group total ekpi"}, axis=1),
    on = "Material Group"
    )
df_country_ekpi_agg["Material Group country share"] = df_country_ekpi_agg["Valued_Ekpi_intensity"] / df_country_ekpi_agg["Material Group total ekpi"]
df_country_ekpi_agg
```

```python
df_country_ekpi_agg[df_country_ekpi_agg["Material Group"] == "Animal Fibers"]
```

```python
df_ekpi_perkg = pd.read_csv("data/material-quantity-in-2019-and-2020.csv", sep=";")
df_ekpi_perkg.rename({"Material group (exclude unspecified)": "material_group"}, axis=1, inplace=True)
df_ekpi_perkg["value_per_kg"] = df_ekpi_perkg["Valued result"]/df_ekpi_perkg["Quantity_kg"]
df_ekpi_perkg
```

```python
df_material_ekpi_agg = df_country_ekpi.groupby(["Material Group", "Environmentalimpactgroup"]).sum().reset_index()
df_material_ekpi_agg.rename({"Material Group": "material_group", "Environmentalimpactgroup": "env_impact_group"}, axis=1, inplace=True)
df_material_ekpi_agg = df_material_ekpi_agg.merge(df_ekpi_perkg)
df_material_ekpi_agg["env_impact_per_kg"] = df_material_ekpi_agg["Material_ePKI_intensity"] / df_material_ekpi_agg["Quantity_kg"]
df_material_ekpi_agg
```
